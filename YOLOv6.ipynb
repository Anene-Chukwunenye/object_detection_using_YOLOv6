# Concept
### YOLOv6 is a single-stage object detection framework dedicated to industrial applications.
### YOLO divides an image into a grid system, and each grid detects objects within itself. It can be used for real-time inference and require very few computational resources.
### YOLOv6-M/L also achieves better accuracy performance than other detectors with a similar inference speed.

#### Object identification using the model

```python
# download the yolov6 weight(yolov6n)

import torch 
model_local = torch.hub.load('.', 'yolov6n', source='local') 
```

```python
from imutils import paths
```

```python
# created a path that leads to the directory of the random objects dataset

path = list(paths.list_images("/content/YOLOv6/data/random_objects"))
```

```python
# looping through the entire images in the random objects dataset

for img in path:
  prediction = model_local.predict(img)
  model_local.show_predict(img)
 
```

##### The model unfortunatly does not correctly identify all the objects within the different images. However, more training will be required to improve the model.
